```python
##funao com o laço de repetição while ou while true com dicionarios com chave produto e valor seu preço
##com o inpút o usuario deve achar se o produto existe na lista, e com isso mostrar o seu produto e valor,
#caso não existe o deve se informar que o produto nao está na lista
#deve se pergunta se o usuario gostaria de continuar ou não, caso seja s.lower o programa deve parar

def lista_de_produtos(lista, produto):
    if produto in lista:
        return f"{produto}: {lista[produto]}"
    else:
        return "nao encontrado"

lista = {"TV": "R$3000.00", "NOTEBOOK": "R$2000.00", "CELULAR": "R$1000.00", "CAMA": "R$300.00"}


while True:
    produto = input("Qual produto você está procurando na lista? ").upper()
    print(lista_de_produtos(lista, produto))
    opcao = input("Deseja sair da lista? Aperte 's' para continuar ou qualquer tecla para sair: ").lower()
    if opcao != "s":
        break

```

```python
#crie uma classe que representa uma pessoa, Ela deve possuir um CPF e um dependente
#onde o dependente é outra pessoa, dependente não é obrigatório
#crie duas instancias pai e filho, e imprima as saídas

#exemplos de saída

#nome: Rodrigo CPF:200.300.400-45 Dependente: None
#nome: Fernando CPF:100.200.300-45 Dependente: Rodrigo

class Pessoa:
    def __init__(self, nome, cpf):
        self.nome = nome
        self.cpf = cpf

    def __str__(self):
        return f"nome: {self.nome} CPF: {self.cpf}"

class Dependente(Pessoa):
    def __init__(self, nome, cpf, dependente=None):
        super().__init__(nome, cpf)
        self.dependente = dependente

    def __str__(self):
        return f"nome: {self.nome} CPF: {self.cpf} Dependente: {self.dependente.nome if self.dependente else 'None'}"

rodrigo = Pessoa(nome="Rodrigo", cpf="200.300.400-45")
fernando = Dependente(nome="Fernando", cpf="100.200.300-45", dependente=rodrigo)

print(rodrigo)
print(fernando)

```

```python
class Veiculo:
    def __init__(self, peso, potencia, rodas):
        self.peso = peso
        self.potencia = potencia
        self.rodas = rodas

    def distancia_percorrida(self):
      return (self.potencia / self.rodas) * 1000

    def __lt__(self, menor):
      return self.distancia_percorrida() < menor.distancia_percorrida()



    def __gt__(self, maior):
      return self.distancia_percorrida() > maior.distancia_percorrida()


    def __str__(self):
        return f"Peso: {self.peso} Potência: {self.potencia} Rodas: {self.rodas} "


class Onibus(Veiculo):
  def __init__(self, peso, potencia, rodas):
    super().__init__(peso, potencia, rodas)

class Carro(Veiculo):
  def __init__(self, peso, potencia, rodas):
    super().__init__(peso, potencia, rodas)

class Moto(Veiculo):
  def __init__(self, peso, potencia, rodas):
    super().__init__(peso, potencia, rodas)


onibus = Onibus(1000, 6, 400)
carro = Carro(300, 4, 100)
moto = Moto(100, 2, 50)



print(onibus)
print(carro)
print(moto)

print(f"Distância percorrida pelo ônibus: {onibus.distancia_percorrida()} km")
print(f"Distância percorrida pelo carro: {carro.distancia_percorrida()} km")
print(f"Distância percorrida pela moto: {moto.distancia_percorrida()} km")

print(f"A menor distância percorrida é: {min(onibus, carro, moto)}")
print(f"A maior distância percorrida é: {max(onibus, carro, moto)}")

veiculos = [onibus, carro, moto]
maior_distancia = max(veiculos)
menor_distancia = min(veiculos)

print(f'A menor distância percorrida é: {menor_distancia.distancia_percorrida()} km')
print(f'A maior distância percorrida é: {maior_distancia.distancia_percorrida()} km')
```

```python
#crie uma classe que represente um numero negativo, use propriedades (@property) para controlar o valor guardado pela classe
#sem deixar que ele fique positivo (O 0 pode ).Além disso crie alguns operadores para comparação e de subtração
#cuide para que nenhum valor positivo surja

class NumeroNegativo:
  def __init__(self, valor):
    self.valor = valor

  @property
  def valor(self):
    return self.__valor

  @valor.setter
  def valor(self, valor):
    if valor < 0:
      self.__valor = -valor
    else:
      self.__valor = valor

  def __str__(self):
    return f"Valor: {self.valor}"

  def __sub__(self, other):
    return self.valor - other.valor

  def __lt__(self, other):
    return self.valor < other.valor

  def __gt__(self, other):
    return self.valor > other.valor

  def __eq__(self, other):
    return self.valor == other.valor

numero1 = NumeroNegativo(10)
numero2 = NumeroNegativo(-5)

print(numero1)
print(numero2)
```

```python
#crie uma função que diga se um objeto é primitivo do python, informando que é sempre passado valor
#Ex: [int, float str, bool] ou se é passado por referencia o valor

def e_primitivo(valor):
  primitivos = [int, float, str, bool]
  return type(valor) in primitivos

print(e_primitivo(10))
print(e_primitivo(10.0))
print(e_primitivo("10"))
print(e_primitivo(True))
```
